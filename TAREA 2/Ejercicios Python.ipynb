```python

import math
#TAREA 1 - EJERCICIOS BASICOS EN PYTHON

# SU NOMBRE:MILTON TARSIAN GARCIA PAYERO
# MATRICULA: 2020-10676 

#Todos los ejercicios deben ser creados en Jupyter

#1.	Declarar variable de los diferentes tipos, asignarles valor e imprimir el valor. 

# Entero
entero = 5
print("Entero:", entero)

# Decimal
decimal = 3.14
print("Decimal:", decimal)

# Cadena de texto
texto = "Hola, mundo!"
print("Texto:", texto)

# Booleano
booleano = True
print("Booleano:", booleano)


#2. Calcular la distancia. Solicitar al usuario ingresar la velocidad y el tiempo 

velocidad = float(input("Ingrese la velocidad (en m/s): "))
tiempo = float(input("Ingrese el tiempo (en segundos): "))

distancia = velocidad * tiempo
print("La distancia es:", distancia, "metros")


#3. Calcular el promedio simple de notas de un estudiante de sustres notas parciales N1, N2, N3

nota1 = float(input("Ingrese la primera nota: "))
nota2 = float(input("Ingrese la segunda nota: "))
nota3 = float(input("Ingrese la tercera nota: "))

promedio = (nota1 + nota2 + nota3) / 3
print("El promedio es:", promedio)


#4. Hacer un programa que permita ingresar el número de partidos ganados, perdidos y empatados,
# por noticias SIN  en  el  torneo  apertura, se  debe  de  mostrar  su  puntaje  total,  
# teniendo  en  cuenta  que  por  cada  partido ganado obtendrá 3 puntos, empatado 1 punto y perdido 0 puntos.

ganados = int(input("Número de partidos ganados: "))
empatados = int(input("Número de partidos empatados: "))
perdidos = int(input("Número de partidos perdidos: "))

puntaje_total = (ganados * 3) + empatados
print("El puntaje total es:", puntaje_total)


#5. Calcular el perimetro y superficie d eun rectangulo. Se debe solicitar al usuario introduzca la base y la altura.

base = float(input("Ingrese la base del rectángulo: "))
altura = float(input("Ingrese la altura del rectángulo: "))

perimetro = 2 * (base + altura)
superficie = base * altura

print("El perímetro es:", perimetro)
print("La superficie es:", superficie)


#6. Calcular el area y el volumen de un cilindro. Soliictar al usuario introduzca el radio y la altura.

radio = float(input("Ingrese el radio del cilindro: "))
altura = float(input("Ingrese la altura del cilindro: "))

area_base = math.pi * radio**2
volumen = area_base * altura

print("El área de la base es:", area_base)
print("El volumen es:", volumen)

#7. Calcular el volumen de una esfera. Pedir al usuario introduzca el Radio.

radio_esfera = float(input("Ingrese el radio de la esfera: "))
volumen_esfera = (4/3) * math.pi * radio_esfera**3

print("El volumen de la esfera es:", volumen_esfera)

#8. Calcular el Perímetro de un tríangulo equilatero. Pedir al usuario ingrese la altura del triangulo.

altura_triangulo = float(input("Ingrese la altura del triángulo equilátero: "))
lado_triangulo = altura_triangulo * math.sqrt(3)

perimetro_triangulo = 3 * lado_triangulo
print("El perímetro del triángulo equilátero es:", perimetro_triangulo)

#9. Calcular el cambio de moneda dominicana a dolares y euros. pedir al usuario introduzca cantidad de pesos.

cantidad_pesos = float(input("Ingrese la cantidad de pesos dominicanos: "))
tasa_dolar = 58.76  # Tasa de cambio a dólares
tasa_euro =  63.77 # Tasa de cambio a euros

cantidad_dolares = cantidad_pesos * tasa_dolar
cantidad_euros = cantidad_pesos * tasa_euro

print("Cantidad en dólares:", cantidad_dolares)
print("Cantidad en euros:", cantidad_euros)


#10. Calcule la velocidad de un automóvil. Pedir al usuario ingrese tiempo en segundos y distincia en metros.

distancia_auto = float(input("Ingrese la distancia recorrida por el automóvil (en metros): "))
tiempo_auto = float(input("Ingrese el tiempo que tardó el automóvil (en segundos): "))

velocidad_auto = distancia_auto / tiempo_auto
print("La velocidad del automóvil es:", velocidad_auto, "m/s")
```
