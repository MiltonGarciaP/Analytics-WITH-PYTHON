```python
#Milton Tarsian Garcia Payero

import pandas as pd

# Cargar los datos a un DataFrame
df = pd.read_csv("dataCGP.csv")

# 2.1 Agrupar los datos por Fecha y producto y arrojar el total
total_por_fecha_producto = df.groupby(['Fecha', 'Producto'])['Total'].sum()
# print(total_por_fecha_producto)
# 2.2 Agrupar los datos por fecha y producto y arrojar la cantidad total
cantidad_por_fecha_producto = df.groupby(['Fecha', 'Producto'])['Cantidad'].sum()
#print(cantidad_por_fecha_producto)
# 2.3 Agrupar los datos por producto y cantidad y sacar la media de cada producto
media_por_producto = df.groupby(['Producto'])['Cantidad'].mean()
#print(media_por_producto)
# 2.4 Agrupar los datos por fecha y cantidad vendida para obtener la media por fecha
media_por_fecha = df.groupby(['Fecha'])['Cantidad'].mean()
#print(media_por_fecha)
# 2.5 Agrupar los datos por fecha, producto y total para obtener la media de los montos totales
media_total_por_fecha_producto = df.groupby(['Fecha', 'Producto'])['Total'].mean()
print(media_total_por_fecha_producto)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 3.1 Grfico de barras
total_por_fecha_producto.unstack().plot(kind='bar', stacked=True)
plt.title('Total por Fecha y Producto')
plt.xlabel('Fecha')
plt.ylabel('Total')
plt.xticks(rotation=45)
plt.show()

# 3.2 Grfico de lneas
media_por_fecha.plot(kind='line')
plt.title('Media de Cantidad Vendida por Fecha')
plt.xlabel('Fecha')
plt.ylabel('Media de Cantidad Vendida')
plt.xticks(rotation=45)
plt.show()

# 3.3 Histogramas
sns.histplot(df['Cantidad'], bins=20, kde=True)
plt.title('Histograma de Cantidad')
plt.xlabel('Cantidad')
plt.ylabel('Frecuencia')
plt.show()

# 3.4 Grfico de pastel
media_total_por_fecha_producto.unstack().plot(kind='pie', subplots=True, figsize=(12, 8), autopct='%1.1f%%')
plt.title('Media de Montos Totales por Fecha y Producto')
plt.ylabel('')
plt.show()
```
